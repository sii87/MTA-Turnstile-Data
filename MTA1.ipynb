```python
#I Created a column of type datetime
df_2021["DATE_TIME"] = pd.to_datetime(df_2021.DATE + " " + df_2021.TIME, format="%m/%d/%Y %H:%M:%S")
df_2021
```

```python
#Now I will creat a new dataframe that contain only Sep (to be used later).
S2021_only = df_2021[(df_2021["DATE_TIME"] >= "2021-9-1") & (df_2021["DATE_TIME"] < "2021-10-1")]
S2021_only.DATE.value_counts().sort_index()
```

```python
#I will CONVER the DATE column to DateTime 
df_2021.DATE = pd.to_datetime(df_2021.DATE)
```

```python
df_2021.drop_duplicates(subset=["C_A", "UNIT", "SCP", "STATION", "DATE_TIME"], inplace=True)
df_2021
```

```python
data_turnstile = S2021_only.copy()

data_turnstile["Per_turnstile"] = S2021_only.C_A + " " + S2021_only.UNIT + " " + S2021_only.SCP
data_turnstile
```

```python
data_turnstile.STATION.value_counts().sort_index()
```

```python
#There are 379 stations in 2021, and i want to find the station that has the maximum number of tuirnstiles.
```

```python
data_turnstile_groupby = (data_turnstile
 .groupby(["STATION", "DATE_TIME"])
 .Per_turnstile.count()
 .reset_index()
 .sort_values("Per_turnstile", ascending=False))


data_top_5_station_turnstile = data_turnstile_groupby.groupby(["STATION"]).Per_turnstile.max().reset_index().sort_values("Per_turnstile", ascending=False).head(5)
data_top_5_station_turnstile
```

```python
#The station that has the maximum number of turnstiles is 34 ST-PENN STA.
```

```python
plt.bar(data_top_5_station_turnstile['STATION'], data_top_5_station_turnstile['Per_turnstile']);
plt.xticks(rotation = 90)
plt.ylabel("Number of Turnstile")
plt.xlabel("Stations name")
plt.title("The Top 5 Stations in Terms of the Number of Turnstiles");

plt.savefig('top_5_turnstiles.png')
```
